This project contains voting data provided by [YouGov](https://www.voterstudygroup.org/publication/2019-voter-survey-full-data-set) which contains information on personal characteristics through answers to survey questions. Most of the surveyed people participated in earlier surveys available on the companies hompage.<br>
This project tough only focusses on the cross-section. Extensions are possible. <br>
Since the survey contains questions and personal characteristics in dimensions greater than 1000 the aim of this project is to twofold: <br>
First to reduce the dimension in order to obtain voter categories determined by a small number of parameters. <br>
Second to check wheter membership to a cretain group can reliable predict voting behavior.

An accompanying pdf file exists that details the survey questions asked. The main challenge of the data set is that all variables are categorical variables ranging from standard questions like family income breackets to whether people consider themselves 'born-again christians'. The analysis was further complicated by a range of missing values.

As a general idea, we wanted to create prediction which factors (answer to survey questions) would predict whether a person votes democrat or republican. Furthermore, we consired having voted in general as a potential outcome variable. And interesting question would have been to analyze whether there exists more heterogeneity among the voters such that there is potential for more than a two party system in the US. Any analysis in that direction turned out to be infeasible for us.

If the survey question has more than 8 possible answers one could specify a range and 98 indicates "skipped"
Otherwise 8 was used to refrence to "skipped"

```r
idx_na <- colSums(is.na(data)) > 0
data <- data[, !idx_na, with = FALSE]
col_names <- colnames(data)
for(i in 2:ncol(data)){
  if(nlevels(data[[col_names[i]]]) > 8 ){
    for(j in 1: nrow(data)){
      if(data[[j,i]] == 8){
        data[[j,i]] <- NA
      }
    }
  }else{
    for(j in 1:nrow(data)){
      if(data[[j,i]] == 98){
        data[[j,i]] <- NA
      }
      
    }
  }
}
```

```r
idx_na <- colSums(is.na(data)) > 0
data <- data[, !idx_na, with = FALSE]
```

# 1.2. Test/Train/Validation Split

Preparation for future analysis. Since training and testing on the same data set will typically result in an overfitting effect we split our data into test and training sets. <br>
Hyperparameters are chosen based on a validation for similar reasons.

```r
set.seed(123) #should have been 42 
n <- nrow(data_new)
p = ncol(data_new) - 1
idx <- sample(seq(1, 3), size = nrow(data_new), replace = TRUE, prob = c(.8, .2, .2))
set_train <- data_new[idx == 1,]
set_test <- data_new[idx == 2,]
set_validation <- data_new[idx == 3,]
```

# 2. LASSO Regression

In order to find out which variables are important to predict voting behaviour, we want to create a Lasso reg. 

```r
#Glmnet Lasso Regression with Cross Validation 
#Lambda Min chosen as option to select the best lambda. 
lambda_lasso <- cv.glmnet(x = data.matrix(set_train[,-1]),
                    y = data.matrix(set_train[,1]), alpha = 1)$lambda.min
```

```r
lambda_lasso
```

Cross-Validation can be used instead of a simple train-test-val split. The default value for glmnet is 10-fold. 
Since we are not interested too much into the analysis, but rather into getting the models to work, we continued to use the splitted dataset. 

glmnet uses the elastic net optimization, a mixture of ridge and lasso regression governed y the mixture coefficient alpha. For lasso alpha has to be one.

```r
plot(cv.glmnet(x = data.matrix(set_train[,-1]),
                    y = data.matrix(set_train[,1]), alpha = 1))
```

```r
model_lasso <- glmnet(x = data.matrix(set_train[,-1]),y = data.matrix(set_train[,1]), data = set_train, 
                      family = "binomial" ,alpha = 1)
```

```r
summary(model_lasso)
```

```r
plotmo::plot_glmnet(x = model_lasso)
```

As one can see from the graphical inspection of the coefficient paths most variables are shrunken to zero, where only a smaller share is set to be zero for the penalization parameter approaching infinity.

```r
coef_lasso <- coef(model_lasso,  s = lambda_lasso)
```
